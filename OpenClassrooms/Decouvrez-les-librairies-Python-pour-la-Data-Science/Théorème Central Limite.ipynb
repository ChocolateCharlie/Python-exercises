```python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
from math import sqrt, pi, exp
```

```python
domain = range(-100, 100)
mu = 0
sigma = 20

f = lambda x : 1/(sqrt(2 * pi * pow(sigma, 2))) * exp(-pow((x - mu), 2) / (2 * pow(sigma, 2)))
y = [f(x) for x in domain]

plot = plt.plot(domain, y)
```

```python
random_matrix = numpy.random.rand(100, 200)
sums = numpy.sum(random_matrix, 0)
```

```python
print("The size of the sums variable is {}."
      .format(sums.shape))
```

```python
plot = plt.scatter(range(200), sums)
```

```python
plot = plt.hist(sums)
```

```python
random_matrix = numpy.random.rand(10000, 100000)
sums = numpy.sum(random_matrix, 0)

plot = plt.hist(sums, bins = 100)
```

```python
print("The empiric mean of the distribution is {}."
      .format(numpy.mean(sums)))
print("The empiric mean of the variable generated by the rand function is {}."
      .format(numpy.mean(numpy.random.rand(100000))))
print("The empiric variance of the distribution is {}."
      .format(numpy.var(sums)))
print("The empiric variance of the variable generated by the rand function is {}."
      .format(numpy.var(numpy.random.rand(100000))))
```
