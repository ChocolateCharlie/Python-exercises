```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

## Chargement des données à l'aide de Pandas

```python
hubble_df = pd.read_csv("hubble_data.csv")
```

## Visualisation des points de données

```python
fig = plt.figure()
ax = plt.axes()

plt.scatter(hubble_df.distance, hubble_df.recession_velocity, color='#8A0808')

plt.title("Vitesse d'éloignement des galaxies données en fonction de leur distance")
ax = ax.set(xlabel = 'Distance',
            ylabel = 'Vitesse d\'éloignement')
```

De prime abord, la relation entre la vitesse d'éloignement des galaxies et leur distance semble linéaire. Par conséquent, il paraît judicieux de tenter d'appréhender ce phénomène en effectuant une régression linéaire.

# Régression linéaire

```python
distance = np.array(hubble_df.distance).reshape(-1, 1)
velocity = np.array(hubble_df.recession_velocity).reshape(-1, 1)
```

```python
regression = LinearRegression().fit(distance, velocity)
```

```python
velocity_pred = regression.predict(distance)
```

```python
fig = plt.figure()
ax = plt.axes()

plt.plot(distance, velocity_pred, color='#04B404')
plt.scatter(hubble_df.distance, hubble_df.recession_velocity, color='#8A0808')

plt.title("Vitesse d'éloignement des galaxies données en fonction de leur distance")
ax = ax.set(xlabel = 'Distance',
            ylabel = 'Vitesse d\'éloignement')

plt.xticks(())
```
