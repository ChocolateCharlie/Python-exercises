```python
from enum import Enum
import numpy
import matplotlib.pyplot
import monti_hall as mh
```

## Analyse rapide des résultats

```python
print("En changeant de porte, le joueur a gagné {} sur 10000 parties"
      .format(numpy.sum(mh.play(mh.Strategy.CHANGE, 10000)), 0))

print("En gardant son choix initial, le joueur a gagné {} sur 10000 parties."
      .format(numpy.sum(mh.play(mh.Strategy.KEEP, 10000)), 0))
```

## Visualisation avec _plot_

```python
plot = matplotlib.pyplot.plot(mh.play(mh.Strategy.CHANGE, 10))
```

```python
plot = matplotlib.pyplot.plot(mh.play(mh.Strategy.CHANGE, 10000))
```

## Visualisation avec _scatter_

```python
plot = matplotlib.pyplot.scatter(range(10000), mh.play(mh.Strategy.CHANGE, 10000))
```

## Visualisation avec _bar_

```python
plot = matplotlib.pyplot.bar([1, 2],
                             [numpy.sum(mh.play(mh.Strategy.CHANGE, 10000), 0), numpy.sum(mh.play(mh.Strategy.KEEP, 10000), 0)],
                             tick_label=["Changer","Garder"])
```

## Visualiser une liste de nombres

```python
change_outcomes = []
keep_outcomes = []
samples =  [1000, 10000, 20000, 50000, 80000, 100000]
for nb_games in samples:
    change_outcomes.append(mh.play(mh.Strategy.CHANGE, nb_games))
    keep_outcomes.append(mh.play(mh.Strategy.KEEP, nb_games))
```

```python
figure = matplotlib.pyplot.figure()
plot = matplotlib.pyplot.scatter(samples, [numpy.sum(x, 0) for x in change_outcomes])
plot = matplotlib.pyplot.scatter(samples, [numpy.sum(x, 0) for x in keep_outcomes])
```
